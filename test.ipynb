```python
# A function that takes a domain like "example.com"
# and runs this shell command:
#   openssl s_client -connect example.com:443 </dev/null 2>&1
# and returns the output as a string
#
# Hint: use subprocess.run()
# Parameters you need: capture_output=True, text=True

import subprocess

def get_tls_raw(domain):
    result = subprocess.run(['openssl', 's_client', '-connect', f'domain:443'], capture_output=True, text=True, input='')
    return result.stdout
```

```python
def get_tls_version(raw_output):
    # loop through each line
    # if the line contains 'Protocol'
    # return the version (the part after the colon, stripped of whitespace)
    # if nothing found, return 'Unknown

    for line in raw_output.split('\n'):
        if 'Protocol' in line:
            return line.split(':')[1].strip()
    return 'Unknown'

```

```python
def scan_domain(domain):
    # 1. call get_tls_raw(domain) to get the output
    # 2. call get_tls_version() on it
    # 3. get the algorithm — hint: look for a line containing 'Server public key'
    #    that line looks like: "Server public key is 2048 bit RSA"
    # 4. call is_quantum_vulnerable() on the algorithm
    # 5. return a dict with keys:
    #    domain, tls_version, algorithm, quantum_vulnerable
    raw = get_tls_raw(domain)
    version = get_tls_version(raw)
    algorithm = 'Unknown' #default
    for line in raw.split('\n'):
        if 'Server public key' in line:
            algorithm = line.split(' ')[-1]
            break
    quantum_vulnerable = is_quantum_vulnerable(algorithm)
    return {'domain': domain, 'tls_version': version, 'algorithm': algorithm, 'quantum_vulnerable': quantum_vulnerable}
```

```python
def expand_key(master_key):
    """
    Expands and returns a list of key matrices for the given master_key.
    """

    # Round constants https://en.wikipedia.org/wiki/AES_key_schedule#Round_constants
    r_con = (
        0x00, 0x01, 0x02, 0x04, 0x08, 0x10, 0x20, 0x40,
        0x80, 0x1B, 0x36, 0x6C, 0xD8, 0xAB, 0x4D, 0x9A,
        0x2F, 0x5E, 0xBC, 0x63, 0xC6, 0x97, 0x35, 0x6A,
        0xD4, 0xB3, 0x7D, 0xFA, 0xEF, 0xC5, 0x91, 0x39,
    )

    # Initialize round keys with raw key material.
    key_columns = bytes2matrix(master_key)
    iteration_size = len(master_key) // 4

    # Each iteration has exactly as many columns as the key material.
    i = 1
    while len(key_columns) < (N_ROUNDS + 1) * 4:
        # Copy previous word.
        word = list(key_columns[-1])

        # Perform schedule_core once every "row".
        if len(key_columns) % iteration_size == 0:
            # Circular shift.
            word.append(word.pop(0))
            # Map to S-BOX.
            word = [s_box[b] for b in word]
            # XOR with first byte of R-CON, since the others bytes of R-CON are 0.
            word[0] ^= r_con[i]
            i += 1
        elif len(master_key) == 32 and len(key_columns) % iteration_size == 4:
            # Run word through S-box in the fourth iteration when using a
            # 256-bit key.
            word = [s_box[b] for b in word]

        # XOR with equivalent word from previous iteration.
        word = bytes(i^j for i, j in zip(word, key_columns[-iteration_size]))
        key_columns.append(word)

    # Group key words in 4x4 byte matrices.
    return [key_columns[4*i : 4*(i+1)] for i in range(len(key_columns) // 4)]
```

**AES Decryption FULL**

```python
N_ROUNDS = 10

key        = b'\xc3,\\\xa6\xb5\x80^\x0c\xdb\x8d\xa5z*\xb6\xfe\\'
ciphertext = b'\xd1O\x14j\xa4+O\xb6\xa1\xc4\x08B)\x8f\x12\xdd'




def decrypt(key, ciphertext):
    round_keys = expand_key(key) # Remember to start from the last round key and work backwards through them when decrypting

    # Convert ciphertext to state matrix
    state = bytes2matrix(ciphertext)
    # Initial add round key step
    state = add_round_key(state, round_keys[N_ROUNDS])
    for i in range(N_ROUNDS - 1, 0, -1):
        state = inv_shift_rows(state)
        state = inv_sub_bytes(state)
        state = add_round_key(state, round_keys[i])
        inv_mix_columns(state)
    # Run final round (skips the InvMixColumns step)
    state = inv_shift_rows(state)
    state = inv_sub_bytes(state)
    state = add_round_key(state, round_keys[0])
    # Convert state matrix to plaintext
    plaintext = matrix2bytes(state)
    return plaintext


print(decrypt(key, ciphertext))
```

**AES Encryption**

```python
def encrypt(key, plaintext):
    round_keys = expand_key(key)
    state = bytes2matrix(plaintext)

    # Initial round key
    state = add_round_key(state, round_keys[0])

    # Main rounds
    for i in range(1, N_ROUNDS):
        state = sub_bytes(state)
        state = shift_rows(state)
        mix_columns(state)              # in-place
        state = add_round_key(state, round_keys[i])

    # Final round — no mix_columns
    state = sub_bytes(state)
    state = shift_rows(state)
    state = add_round_key(state, round_keys[N_ROUNDS])

    return matrix2bytes(state)

print(encrypt(key, decrypt(key,ciphertext)))
```

**Eliptic Curves**

```python
p = 9739
a = 497

def add_points(P, Q):
    if P is None: return Q
    if Q is None: return P
    x1, y1 = P
    x2, y2 = Q
    if x1 == x2 and y1 == (-y2 % p):
        return None
    if P == Q:
        lam = (3*x1**2 + a) * pow(2*y1, -1, p) % p
    else:
        lam = (y2 - y1) * pow(x2 - x1, -1, p) % p
    x3 = (lam**2 - x1 - x2) % p
    y3 = (lam*(x1 - x3) - y1) % p
    return (x3, y3)

```

```python
def scalar_mul(p,n):
  # return n * p
  P = p
  Q = p
  R=(0,0) # Identity
  while n>0:
    if n&1:
      R=add_points(R,Q)
    Q=add_points(Q,Q)
    n=n//2
  return R   
```

```python
import json
import hashlib
from pwn import remote
from ecdsa.ecdsa import generator_192, Public_key, Private_key, Signature
from Crypto.Util.number import bytes_to_long

g = generator_192
n = g.order()

def sha1(data):
    return hashlib.sha1(data).digest()

r = remote('socket.cryptohack.org', 13381)
r.recvline()  # banner

# Step 1: Get a signed timestamp - nonce k is only 1-59!
r.sendline(json.dumps({"option": "sign_time"}).encode())
resp = json.loads(r.recvline())
print("Got:", resp)

msg = resp['msg']
sig_r = int(resp['r'], 16)
sig_s = int(resp['s'], 16)

hsh = bytes_to_long(sha1(msg.encode()))

# Step 2: Brute force k (only 1-59 possible since seconds value reused as range)
found_k = None
found_privkey = None
for k in range(1, 60):
    # ECDSA: s = k^-1 * (hash + r*privkey) mod n
    # => privkey = (s*k - hash) * r^-1 mod n
    try:
        r_inv = pow(sig_r, -1, n)
        privkey_candidate = ((sig_s * k - hsh) * r_inv) % n
        # verify by checking r matches: R = k*G, r = R.x() mod n
        R = k * g
        if R.x() % n == sig_r:
            found_k = k
            found_privkey = privkey_candidate
            break
    except Exception as e:
        continue

print("Found k =", found_k)
print("Recovered private key =", found_privkey)

# Step 3: Sign "unlock" with recovered private key
pubkey = Public_key(g, g * found_privkey)
privkey = Private_key(pubkey, found_privkey)

unlock_hash = bytes_to_long(sha1(b"unlock"))
sig = privkey.sign(unlock_hash, 12345)  # any random nonce works now

payload = {
    "option": "verify",
    "msg": "unlock",
    "r": hex(sig.r),
    "s": hex(sig.s)
}

r.sendline(json.dumps(payload).encode())
print(r.recvline())
```
